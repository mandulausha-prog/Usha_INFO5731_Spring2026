```python
import requests

url = "https://remoteok.com/api"
headers = {"User-Agent": "Mozilla/5.0"}

data = requests.get(url, headers=headers).json()

# First item is metadata, remove it
jobs = [job for job in data if isinstance(job, dict) and job.get("id") not in (None, 0)]

print("Total job postings:", len(jobs))










```

```python
from collections import Counter

companies = [job.get("company") for job in jobs if job.get("company")]
top_3_companies = Counter(companies).most_common(3)

print("Top 3 companies:")
top_3_companies

```

```python
count = 0

for job in jobs:
    title = job.get("position") or job.get("title", "")
    if "data" in title.lower():
        count += 1

print('Job titles containing "Data":', count)


```
