## Q1 (1 point) — Answer below

```python
import pandas as pd

df = pd.read_csv("support_messages.txt", sep="\t")

print(df.shape)
print(df.head(3))

```

## Q2 (3 points) — Answer below

```python
df["word_count"] = df["message"].apply(lambda x: len(str(x).split()))
df["char_count"] = df["message"].apply(lambda x: len(str(x)))
df["num_count"] = df["message"].apply(lambda x: sum(c.isdigit() for c in str(x)))
df["upper_word_count"] = df["message"].apply(lambda x: sum(word.isupper() for word in str(x).split()))

print(df)

```

## Q3 (3 points) — Answer below

```python
import re
from nltk.corpus import stopwords
import nltk

nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    text = " ".join(words)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df["clean"] = df["message"].apply(clean_text)

print(df.loc[df["id"].isin([1,4]), ["message", "clean"]])
```

## Q4 (2 points) — Answer below

```python
df["order_id"] = df["message"].str.extract(r'(ord-\d+)', flags=re.IGNORECASE)
df["email"] = df["message"].str.extract(r'(\S+@\S+)')

print(df[["id", "order_id", "email"]])

```

## Q5 (1 point) — Answer below

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["clean"])

feature_names = np.array(vectorizer.get_feature_names_out())
avg_tfidf = np.asarray(X.mean(axis=0)).ravel()

top_indices = avg_tfidf.argsort()[-5:][::-1]
top_keywords = feature_names[top_indices]

print(top_keywords)

```
